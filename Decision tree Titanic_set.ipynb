```python
#filling categorical variable with dummy values
df1=pd.get_dummies(df1)
df1
```

```python
x=df1.iloc[:,2:]
y=df1.iloc[:,1]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
# Build model
```

```python
from sklearn.tree import DecisionTreeClassifier
model_1=DecisionTreeClassifier().fit(x_train,y_train)
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
#Evaluation of Model
pd1=model_1.predict(x_test)
acc=accuracy_score(y_test,pd1)
cm=confusion_matrix(y_test,pd1)
cr=classification_report(y_test,pd1)
print('Acc=',acc)
print(sns.heatmap(cm,annot=True))
print('CR=',cr)
```

```python
# Hyperparameter
criterion = ["gini" , "entropy"]
max_depth = list(range(1,11))
grid = {"criterion":criterion , "max_depth":max_depth}
```

```python
# Build model
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
cv = RepeatedStratifiedKFold(n_splits=11 , n_repeats=3 , random_state=1)

grid_cv = GridSearchCV(DecisionTreeClassifier() , param_grid= grid , cv=cv , scoring="accuracy")

res = grid_cv.fit(x_train, y_train)
print(res.best_score_)
print(res.best_params_)
```

```python
model_2=DecisionTreeClassifier(criterion ='entropy', max_depth=3).fit(x_train , y_train)
pd2 = model_2.predict(x_test)
pd2
```

```python
# Evaluation of model
acc = accuracy_score(y_test , pd2)
cm = confusion_matrix(y_test , pd2)
cr = classification_report(y_test , pd2)
print('Acc=' ,acc)
print(sns.heatmap(cm, annot=True))
print("CR=" , cr)
```
